```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
sns.set_style('darkgrid')
```

```python
df1 = pd.read_csv('./PPW-vs-PPP-2010.csv')
print(df1.columns)
df1.head()
```

```python
df1['Total_waste(kg/day)'] = df1['PPW(kg/person/day)'] * df1['Total Population']
df1['Total_waste(kg/year)'] = df1['Total_waste(kg/day)'] * 365
df1.head()
```

```python
df2 = pd.read_csv('./PMPW-vs-PPP-vs-TotalPopulation-2010.csv')
print(df2.columns)
df2.head()
```

<hr style="border-top: dotted 15px red;background-color:#0000">

```python
## "scatter plot graph" 
## Finding any corelation between GDP percapita and Mismanaged waste per person(kg/day)
plt.figure(1, figsize=(12,8))

plt.scatter(df2['PPP'], df2['PMPW'])
sns.regplot(x='PPP', y='PMPW', data=df2,scatter_kws={'color': '#34568B'}, line_kws={'color': '#650021'})

plt.title('Waste Mismanaged', loc='center', fontsize=15)
plt.ylabel('Mismanaged waste per person(kg/day)', loc='center', fontsize=15)
plt.xlabel('GDP per capita', fontsize=12)
plt.show()
```

```python
## "scatter plot graph" 
## Finding any corelation between "GDP-percapita" and "Waste generated perperson"
plt.figure(1, figsize=(12,8))

plt.scatter(df1['PPP'], df1['PPW(kg/person/day)'])
sns.regplot(
    x='PPP', y='PPW(kg/person/day)', data=df1,
    scatter_kws={'color': '#CD212A'}, 
    line_kws={'color': '#380282'}
)

plt.title('Waste Generated by GDP', loc='center', fontsize=15)
plt.ylabel('Waste per person (kg/day)', loc='center', fontsize=15)
plt.xlabel('GDP per capita', fontsize=12)
plt.show()
```
